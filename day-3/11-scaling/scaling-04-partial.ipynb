---------------------------
### Partial scaling 
- on breast cancer dataset
-----------------

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler

from sklearn import datasets
```

```python
data = datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(data = data.data, columns = data.feature_names )
```

```python
df.sample(5)
```

```python
df.columns
```

```python
ct = ColumnTransformer(
        [ ('std-sclr',      StandardScaler(),  ['worst concave points', 'worst symmetry', 'worst fractal dimension']),
          ('min-ax',        MinMaxScaler(),    ['smoothness error', 'compactness error', 'concavity error']),
          ('robust_scaler', RobustScaler(),    ['mean smoothness', 'mean compactness', 'mean concavity'])
        ],
        remainder='passthrough'
    )
```

```python
scaled_X = ct.fit_transform(df)
```

```python
scaled_X
```

```python
scaled_X_df = pd.DataFrame(scaled_X, columns=df.columns)
```
