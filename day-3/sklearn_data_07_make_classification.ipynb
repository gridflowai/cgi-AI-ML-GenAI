---------------------------------
#### make_classification 
- is a function in the sklearn.datasets module in scikit-learn that generates a random classification dataset
-------------------------------------------

- `n_samples` (int): The total number of samples.
- `n_features` (int): The total number of features.
- `n_informative` (int, optional, default=2): The number of informative features.
- `n_redundant` (int, optional, default=2): The number of redundant features.
- `n_classes` (int, optional, default=2): The number of classes.
- `n_clusters_per_class` (int, optional, default=2): The number of clusters per class.
- `weights` (list or tuple of floats, optional, default=None): The proportions of samples assigned to each class.
- `random_state` (int, RandomState instance or None, optional, default=None): Seed or RandomState for reproducibility.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
# Generate a classification dataset
X, y = make_classification(n_samples    = 1000, 
                           n_features   = 2, 
                           n_informative= 2, 
                           n_redundant  = 0,
                           class_sep    = 2.0,
                           n_clusters_per_class=1, 
                           weights      =[0.5, 0.5], 
                           random_state =42)
```

```python
# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k', marker='o', s=50)
plt.title('Generated Classification Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
