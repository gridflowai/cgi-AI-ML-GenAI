#### Imputation with Mean/Median/Mode:

- Replace missing values with the mean, median, or mode of the respective column.

```python
mean_age = titanic['Age'].mean()
titanic['Age'].fillna(mean_age, inplace=True)

mode_cabin = titanic['Cabin'].mode()
titanic['Cabin'].fillna(mode_cabin, inplace=True)

mode_embarked = titanic['Embarked'].mode()
titanic['Embarked'].fillna(mode_embarked, inplace=True)
```
