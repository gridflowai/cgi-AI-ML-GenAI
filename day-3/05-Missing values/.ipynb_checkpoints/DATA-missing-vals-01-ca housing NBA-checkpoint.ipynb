1. longitude: A measure of how far west a house is; a higher value is farther west
2. latitude: A measure of how far north a house is; a higher value is farther north
3. housing_median_age: Median age of a house within a block; a lower number is a newer building
4. total_rooms: Total number of rooms within a block
5. total_bedrooms: Total number of bedrooms within a block
6. population: Total number of people residing within a block
7. households: Total number of households, a group of people residing within a home unit, for a block
8. median_income: Median income for households within a block of houses (measured in tens of thousands of US Dollars)
9. ocean_proximity: Location of the house w.r.t ocean/sea
10. median_house_value: Median house value for households within a block (measured in US Dollars)

#### Handle missing values

```python
# find the count of missing values
df.isnull().sum()
```

```python
# fill missing values with mean column values
df.fillna(df.mean(), inplace=True)

# count the number of NaN values in each column
print(df.isnull().sum())
```

#### Method - 2

```python
df = pd.read_csv(location)
```

```python
# count the number of NaN values in each column
print(df.isnull().sum())
```

```python
# statistics of "total_bedrooms " feature
df['total_bedrooms'].describe()
```

statistics for missing value of "total_bedrooms " feature

    - average is 537 for total bedrooms
    - first quartile is highly low compared to other two

```python
data_null=df[df['total_bedrooms'].isnull()]
data_null.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 8))

plt.scatter(x=df['longitude'], 
            y=df['latitude'], 
            alpha=.3, 
            label='Households')

plt.scatter(x=data_null['longitude'], 
            y=data_null['latitude'], 
            c='r', 
            s=3, 
            label = 'null total_bedrooms')

plt.title("Distribution of households", size=15)
plt.legend();
```

From above plot we can infer that,

- Housing prices are related to location and population density.
- Housing prices near ocean are higher except in northern california.

```python
# making data frame from csv file 
nba = pd.read_csv(location) 
```

```python
nba.isnull().sum()
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
nba['Salary'].describe()
```

```python
sns.kdeplot(nba['Salary']);
```

`Observation`

- Salary column
    - 70% of the players get less than 650,000
    - few players get more than 1M
    
- For missing value, treatment
    - mean
        - missing value will be impacted by the extreme values
    - median
        - seems like a better option
    - MIN
        - not a good option
    - MAX 
        - not a good option


```python
# Or, depending on your needs:
nba['Salary'] = nba['Salary'].fillna(nba['Salary'].mean())
nba['Salary'] = nba['Salary'].fillna(nba['Salary'].min())
nba['Salary'] = nba['Salary'].fillna(nba['Salary'].max())
```

#### when should we use median()
