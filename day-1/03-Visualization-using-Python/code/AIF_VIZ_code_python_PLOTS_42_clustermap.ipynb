```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights.dtypes
```

```python
flights['year'] = flights.year.astype('category')
```

```python
flights = flights.pivot('month', 'year', 'passengers')
```

```python
flights
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights, figsize=(8, 6));
```

```python
sns.clustermap(flights, col_cluster=False)
```

```python
sns.clustermap(flights, row_cluster=False)
```

```python
sns.clustermap(flights, cmap='coolwarm')
```

```python
sns.clustermap(flights, cmap='Blues_r', linewidth=1);
```

```python
sns.clustermap(flights, cmap='Blues_r', linewidth=1, figsize=(8, 6));
```
