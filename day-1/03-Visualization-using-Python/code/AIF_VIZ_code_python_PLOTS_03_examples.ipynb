```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']  = (6, 4)
plt.rcParams['axes.grid']       = True
plt.rcParams['lines.linewidth'] = 2.0
```

``Line plots``

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot');
```

``Scatter Plot``

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot');
```

``Bar plots``

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 15, 7, 12, 9]

plt.bar(x, y)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot');
```

``Histograms``

```python
data = [1, 2, 3, 3, 4, 4, 4, 5, 5, 5, 5]

plt.hist(data, bins=5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram');
```

``Pie Chart``

```python
sizes = [15, 30, 25, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart');
```

```python
data = [10, 15, 7, 12, 9, 20, 5, 8]

plt.boxplot(data)
plt.ylabel('Values')
plt.title('Box Plot');
```

``area plot``

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.fill_between(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Plot');
```

```python
days = [1, 2, 3, 4, 5]

work     = [6, 7, 8, 6, 7]
study    = [2, 3, 2, 3, 4]
exercise = [1, 2, 1, 1, 2]

plt.stackplot(days, work, study, exercise, labels=['Work', 'Study', 'Exercise'])
plt.xlabel('Days')
plt.ylabel('Hours')

plt.title('Stack Plot')
plt.legend(loc='upper left');
```

``Stacked Bar Chart``

```python
x = ['A', 'B', 'C', 'D', 'E']
y1 = [10, 15, 7, 12, 9]
y2 = [5, 8, 4, 10, 6]

plt.bar(x, y1, label='Y1')
plt.bar(x, y2, bottom=y1, label='Y2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend();
```

``heatmap``

```python
data = np.random.rand(15, 15)

plt.imshow(data)
plt.colorbar()
plt.title('Heatmap');
```

``3D Surface Plot``

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Surface Plot');
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Generate random data for the 3D scatter plot
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)

# Create the 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c='b', marker='o', s=30)

# Function to update the view for each animation frame
def update(frame):
    ax.view_init(elev=20, azim=frame)

# Create the animation
animation = FuncAnimation(fig, update, frames=np.arange(0, 360, 2), interval=50)

# Show the animation
plt.show()
```

``Error Bar Plot``

```python
x     = [1,  2,  3,  4,  5]
y     = [10, 15, 7,  12, 9]
error = [1,  2, 1.5, 0.5, 1]

plt.errorbar(x, y, yerr=error, fmt='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Error Bar Plot');
```

``Bar Plot with Error Bars``

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 39, 16, 25]
error = [0.5, 1, 1.5, 2, 2.5]

plt.bar(x, y, yerr=error, capsize=5);
```

``Contour Plot``

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Contour Plot');
```

``Contourf Plot``

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar();
```

``Multiple Subplots``

```python
x =  [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [0, 2, 4, 6, 8]

fig, axs = plt.subplots(nrows= 1, ncols=2, sharex=True, figsize=(18, 5))
axs[0].plot(x, y1)
axs[1].plot(x, y2);
```

```python
from ipywidgets import interactive, FloatSlider

x = np.linspace(0, 10, 100)

def plot_line(a, b):
    y = a * x + b
    plt.plot(x, y)
    plt.ylim(0, 100)
    plt.show()

interactive_plot = interactive(plot_line, a=FloatSlider(min=0, max=10, step=0.1, value=1), b=FloatSlider(min=0, max=10, step=0.1, value=0))
interactive_plot
```
