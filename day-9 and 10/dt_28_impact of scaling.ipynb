```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
```

```python
# Load the PIMA dataset with proper column names
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
column_names = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
pima = pd.read_csv(url, header=None, names=column_names)
X = pima.iloc[:, :-1]
y = pima.iloc[:, -1]
```

```python
# Train a decision tree classifier without scaling
clf = DecisionTreeClassifier(random_state=42, max_depth=2)
clf.fit(X, y)
```

```python
# Plot the decision tree without scaling
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=X.columns, rounded=True, class_names=['No Diabetes', 'Diabetes'])
plt.title("Decision Tree without Scaling")
plt.show()
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train a decision tree classifier on the scaled features
clf_scaled = DecisionTreeClassifier(max_depth=2, random_state=42)
clf_scaled.fit(X_scaled, y)
```

```python
# Plot the decision tree with scaling
plt.figure(figsize=(20,10))
plot_tree(clf_scaled, filled=True, feature_names=X.columns, rounded=True, class_names=['No Diabetes', 'Diabetes'])
plt.title("Decision Tree with Scaling")
plt.show()
```
