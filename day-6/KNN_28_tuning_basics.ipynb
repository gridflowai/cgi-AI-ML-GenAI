```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating example data with two features
np.random.seed(42)
X = 5 * np.random.rand(200, 2)
y = (X[:, 0] + X[:, 1] > 4).astype(int)
```

```python
# Applying KNN with k=1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plotting the original data points by class
scatter = axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k', marker='o', label='True Data')
axes[0].set_title('Original Data Points by Class')
axes[0].set_xlabel('Beak Length')
axes[0].set_ylabel('Voice Pitch')
axes[0].legend()

# Adding a colorbar to the first subplot
#cbar = fig.colorbar(scatter, ax=axes[0], label='Class')

# Creating a meshgrid for decision boundary
xx, yy = np.meshgrid(np.linspace(0, 5, 100), np.linspace(0, 5, 100))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotting the KNN decision boundary
contour = axes[1].contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
axes[1].scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k', marker='o', label='True Data')
axes[1].set_title('KNN Decision Boundary with Two Features (k=1)')
axes[1].set_xlabel('Beak Length')
axes[1].set_ylabel('Voice Pitch')
axes[1].legend()

# Adding a colorbar to the second subplot
#cbar2 = fig.colorbar(contour, ax=axes[1], label='Class')

# Adjust layout for better spacing
plt.tight_layout()
plt.show();
```
