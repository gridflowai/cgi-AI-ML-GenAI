----------------------
#### generate a random sample from a continuous uniform distribution 

- within a specified range and plot its probability density function (PDF) 
-----------------------

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the lower and upper bounds of the uniform distribution
a = 2  # Lower bound
b = 6  # Upper bound
```

```python
# Generate a random sample of 1000 values from the uniform distribution
sample_size  = 1000
uniform_data = np.random.uniform(a, b, sample_size)
```

```python
# Create a histogram to visualize the data distribution
plt.hist(uniform_data, bins=20, density=True, alpha=0.6, color='g', label='Sample Data');
```

```python
# Calculate the theoretical PDF of the uniform distribution
x   = np.linspace(a, b, 1000)
pdf = np.full_like(x, 1 / (b - a))  # Probability density function
plt.plot(x, pdf, 'r', lw=2, label=f'PDF: 1 / ({b} - {a})')

# Set plot labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Continuous Uniform Distribution')
plt.legend()

# Show the plot
plt.grid()
plt.show()
```
