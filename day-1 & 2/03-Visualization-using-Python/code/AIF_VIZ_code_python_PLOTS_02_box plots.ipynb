## Box Plots
-------------------

- `Boxplot` is probably one of the most common type of graphic. 

- It gives a nice summary of one or several numeric variables. 

    - The line that divides the box into 2 parts represents the median of the data. 
    - The end of the box shows the `upper` and `lower` quartiles. 
    - The extreme lines shows the `highest` and `lowest` value `excluding outliers`. Note that boxplot hide the number of values

existing behind the variable. Thus, it is highly advised to print the number of observation, add unique observation with jitter or use a violinplot if you have many observations.

#### basic idea on box plots

generate n random integers

```python
n = 15
arr_of_values = np.random.choice(a=np.arange(start=10, stop=50), size=n, replace=False)
```

```python
arr_of_values.sort()
arr_of_values
```

plot the values

```python
plt.figure(figsize=(18, 4))

plt.grid()

plt.axhline(y=0, color='red');

plt.scatter(x= arr_of_values, 
            y= [0]*len(arr_of_values), 
            s= 3 * arr_of_values, 
            zorder=1, 
            alpha = 0.5,
            color ='green')

plt.xticks(range(0, 50))
plt.yticks([-2, -1, 0, 1, 2])
plt.title('Number of values in the array = {}, Min = {}, Max = {}'.format(len(arr_of_values), np.min(arr_of_values), np.max(arr_of_values)));
```

$$\text{First Quartile (Q1)} = \frac{1.(n + 1)}{4}\text{th Term} $$

$$\text{First Quartile (Q2)} = \frac{2.(n + 1)}{4}\text{th Term} $$

$$\text{First Quartile (Q3)} = \frac{3.(n + 1)}{4}\text{th Term} $$

define a function to return quartile term and value of that term

```python
def compute_qn(arr_of_values, qn=1):
    q_val_term = qn * (len(arr_of_values)/4)
    
    if q_val_term% np.floor(q_val_term) ==0:
        return arr_of_values[np.floor(q_val_term).astype('int32') -1], np.floor(q_val_term).astype('int32')
    else:
        terms = [np.floor(q_val_term).astype('int32'), np.ceil(q_val_term).astype('int32')]
        floor_val = arr_of_values[np.floor(q_val_term).astype('int32') -1]
        ceil_val  = arr_of_values[np.ceil(q_val_term).astype('int32')  -1]
        
        return (floor_val + ceil_val)/2, terms
```

lets calculate the quartiles Q1, Q2, Q3 ...

```python
q1, q1_terms = compute_qn(arr_of_values, 1)
q2, q2_terms = compute_qn(arr_of_values, 2)
q3, q3_terms = compute_qn(arr_of_values, 3)
```

```python
q1, q1_terms
```

```python
q2, q2_terms
```

```python
q3, q3_terms
```

```python
plt.figure(figsize=(18, 6))
plt.grid()

plt.axhline(y=0, color='red')

plt.axvline(q1, color='blue')    # lower quartile, Q1
plt.text(x=q1+.2, y=0.1, s='Q1, terms : {}'.format(q1_terms), rotation=90, fontsize=8)

plt.axvline(q2, color='yellow')  # median, Q2
plt.text(x=q2+.2, y=0.1, s='Q2, terms : {}'.format(q2_terms), rotation=90, fontsize=8)

plt.axvline(q3, color='cyan')    # upper quartile, Q3
plt.text(x=q3+.2, y=0.1, s='Q3, terms : {}'.format(q3_terms), rotation=90, fontsize=8)

plt.scatter(x= arr_of_values, 
            y= [0] * len(arr_of_values), 
            s= 3   * arr_of_values, 
            zorder=1, 
            alpha = 0.5,
            color ='green')

plt.xticks(range(50))
plt.yticks([-2, -1, 0, 1, 2])
plt.title('Number of values in the array = {}, Min = {}, Max = {}'.format(len(arr_of_values), np.min(arr_of_values), np.max(arr_of_values)));
```

```python
plt.figure(figsize=(18, 6))
plt.grid()

plt.axhline(y=0, color='red', )

#plt.axvline(q1, color='blue')    # lower quartile, Q1
plt.text(x=q1+.2, y=1.2, s='Q1, terms : {}'.format(q1_terms), rotation=90, fontsize=8)

#plt.axvline(q2, color='yellow')  # median, Q2
plt.text(x=q2+.2, y=1.2, s='Q2, terms : {}'.format(q2_terms), rotation=90, fontsize=8)

#plt.axvline(q3, color='cyan')    # upper quartile, Q3
plt.text(x=q3+.2, y=1.2, s='Q3, terms : {}'.format(q3_terms), rotation=90, fontsize=8)

plt.plot((q1, q3), (2,  2), 'b--')
plt.plot((q1, q3), (1,  1), 'b--')

plt.plot((q2, q2), (1,  2), 'b--')

plt.plot((q1, q1), (1,  2), 'b--')
plt.plot((q3, q3), (1,  2), 'b--')

plt.scatter(x= arr_of_values, 
            y= [0] * len(arr_of_values), 
            s= 3   * arr_of_values, 
            zorder=1, 
            alpha = 0.5,
            color ='green')

plt.xticks(range(50))
plt.yticks([-1, 0, 1, 2, 3])

plt.title('Number of values in the array = {}, Min = {}, Max = {}'.format(len(arr_of_values), np.min(arr_of_values), np.max(arr_of_values)));
```
